```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.collections import PatchCollection
```

```python
%matplotlib notebook
```

```python
fig, ax = plt.subplots()
```

```python
N = 10000
M = 100
dx = 0.05
dy = 0.025
patches = []
for i in range(N):
    patches.append(Rectangle(np.random.random(2), dx, dy))
p = PatchCollection(patches, alpha=0.4)
colors=np.random.random([M, N])
p.set_array(np.array(colors[0, :]))
ax.add_collection(p)
fig.colorbar(p, ax=ax)
```

```python
import ipywidgets as w
sl = w.IntSlider(max=M-1)
def update_slice(change):
    p.set_array(colors[change["new"], :])
    fig.canvas.draw_idle()
sl.observe(update_slice, names="value")
sl
```
